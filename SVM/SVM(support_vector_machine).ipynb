#### we will try to classifer an animal wether it is an crocodile or aligator ?

- size  crocodile are bigger than aligator
- snoth width crocodile have narrow snout and aligator have wider snout

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
# we create 40 separable points
X,y = make_blobs(n_samples=40 , centers = 2 , random_state=20)
```

```python
# fitting the model
clf = svm.SVC(kernel='linear' , C=1)
clf.fit(X,y)
```

```python
# displaying the data
plt.scatter(X[:,0] , X[:,1] , c=y,s=30,cmap=plt.cm.Paired)
plt.xlabel("Width of the snout")
plt.ylabel("Size of the body")
plt.show()
```

```python
## using to predict unknown data
newdata = [[3,4],[5,6]]
print(clf.predict(newdata))
## 0 - crocodile and 1 - aligator
```

```python
clf = svm.SVC(kernel='linear' , C=1)
clf.fit(X,y)
plt.scatter(X[:,0] , X[:,1] , c=y,s=30,cmap=plt.cm.Paired)

ax = plt.gca()    #used to get the current Axes instance on the current figure matching the given keyword
xlim = ax.get_xlim() # used to get the x-axis view limits
ylim = ax.get_ylim() #  used to get the y-axis view limits

# creating grid to evaluate the model
xx = np.linspace(xlim[0] , xlim[1],30)
yy = np.linspace(ylim[0] , ylim[1],30)
YY , XX = np.meshgrid(yy , xx)  
xy = np.vstack([XX.ravel() , YY.ravel()]).T      #ravel() function is used to create a contiguous flattened array
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX,YY,Z,colors='k',levels=[-1,0,1],alpha=0.5,linestyles=['--','-','--'])

#plot support vector
ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100,linewidth=1,facecolors='none')
plt.show()
```
